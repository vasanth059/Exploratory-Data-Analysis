```python
person = data.drop(['Age','Cabin','Lifeboat','Body','Survived'],axis=1)
```

```python
person.head()
```

Data Visualisation

```python
corelation = person.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
sns.pairplot(person)
```

```python
sns.distplot(person['Pclass'],bins=5)
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=person,x='Pclass',y='Parch')
plt.show()
```
